```python
# Apply PCA for X
```

```python
from mlxtend.preprocessing import standardize
from sklearn.decomposition import PCA
```

```python
X = standardize(X)
X
```

```python
pca = PCA(n_components=0.95)
pca.fit(X)

X_pca = pca.transform(X)
X_pca
```

```python
X_pca_df = pd.DataFrame(X_pca)
X_pca_df
```

```python
# Create train and test split sets for X and y variables
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pca_df, y_df, test_size=0.3,random_state=0)
```

```python
model = XGBRegressor(min_child_weight=1, max_depth=5, max_delta_step=0, alpha=0, scale_pos_weight=1, learning_rate=0.1, n_estimators=100, verbosity=1, eta=0.3, objective='reg:squarederror', silent=0, booster='gbtree', tree_method='auto', n_jobs=1, gamma=0)
model.fit(X_train, y_train)
```
